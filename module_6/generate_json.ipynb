```python
import os
import json
from sklearn.model_selection import KFold
```

```python
data_dir = '/home/ikboljon.sobirov/data/fs1_research/Ikboljon.Sobirov/imagecas/imagecas/resampled_space/'
patients = os.listdir(data_dir)
print(len(patients))
```

```python
train_data = []
num_folds = 5

full_idxs = list(range(len(patients)))
kf = KFold(n_splits=num_folds, shuffle=True, random_state=42)

for f in range(num_folds):

    train_idx = {}
    test_idx = {}

    key = 1
    for i,j in kf.split(full_idxs):
        train_idx[key] = i
        test_idx[key] = j

        key += 1

    patients_test = [patients[i] for i in test_idx[f+1]]
    print(patients_test, len(patients_test))

    for p in patients_test:
        train_json_dict = {}
        train_json_dict['patient'] = p
        train_json_dict['fold'] = f
        train_json_dict['image_path'] = os.path.join(data_dir, p, (p+'_img.nii.gz'))
        train_json_dict['mask_path'] = os.path.join(data_dir, p, (p+'_seg.nii.gz'))

        train_data.append(train_json_dict)
```
