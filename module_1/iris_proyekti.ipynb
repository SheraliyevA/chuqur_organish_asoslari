```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
```

### ma'lumotlarini tahlil qilish

```python
iris_df = pd.DataFrame(iris.data, columns=iris.feature_names)
iris_df
```

```python
iris_df['turlari'] = pd.Categorical.from_codes(iris.target, iris.target_names)
```

```python
iris_df
```

```python
grouped = iris_df.groupby('turlari')
grouped.mean()
```

```python
sns.pairplot(iris_df, hue='turlari')
plt.show()
```

```python
# correlation heatmap
sns.heatmap(iris_df[['sepal length (cm)', 'sepal width (cm)', 'petal length (cm)','petal width (cm)']].corr(), annot=True)
plt.show()
```
